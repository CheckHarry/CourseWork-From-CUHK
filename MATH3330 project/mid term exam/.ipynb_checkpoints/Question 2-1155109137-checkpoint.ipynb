```python
import numpy as np
import matplotlib.pyplot as plt
X=np.array([[1,1,1,1,1,1,1,1],[1.6,2.1,3.4,3.9,4.3,5.2,6.1,7.4],[0.5,-0.8,0.2,-0.5,1.2,0.7,0.4,-0.5]]).T
Y=np.array([[3],[2.4],[4.7],[4.5],[6.6],[7.1],[7.3],[8.1]])
Q=2*X.T@X
b=(2*X.T@Y)
print('Q=',Q)
print('b=',b)
```

```python
theta=np.array([[1],[1],[5]])
d=b-Q@theta
g=-d
G=g
for i in range(1,3):
    alpha=-(g.T@d)/(d.T@Q@d)
    theta=theta+alpha*d
    g=Q@theta-b
    beta=(g.T@Q@d)/(d.T@Q@d)
    d=-g+beta*d
         
alpha=-(g.T@d)/(d.T@Q@d)
theta=theta+alpha*d
g=Q@theta-b
```

```python
print(theta)
```

```python
from mpl_toolkits.mplot3d import Axes3D
x = np.arange(0,10.01,0.01)
y = np.arange(0,10.01,0.01)
x1, x2 = np.meshgrid(x,y)
value_predict= theta[0]+theta[1]*x1+theta[2]*x2
figure=plt.figure()
ax=Axes3D(figure)
ax.plot_surface(x1,x2,value_predict)
```
